```python
pt = pd.read_csv('02-7-pop-past-future-2.7-necessities.csv')
pt.tail()
```

```python
ptkeys=ptkeys[1:8]
ptkeys
```

```python
plt.figure(figsize=(16,9))
for i, k in enumerate(ptkeys):
    plt.plot(pt.Year, pt[k], lw=3, label=k)
    loc = i*10
    plt.text(x=pt.Year[loc], y=pt[k][loc], s=k, )
plt.legend();
plt.savefig('figure-2.7a.png')
```

```python
plt.figure(figsize=(16,9))
for i, k in enumerate(ptkeys):
    plt.plot(pt.Year, pt[k], lw=3, label=k)
    loc = i*10
    plt.text(x=pt.Year[loc], y=pt[k][loc], s=k, )
plt.legend();
plt.yscale('log')
plt.ylabel('Population (Million)')
plt.savefig('figure-2.7b-log10.png')
```
