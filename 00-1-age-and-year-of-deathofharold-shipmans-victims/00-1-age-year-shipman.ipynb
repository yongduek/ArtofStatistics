```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
print(sns.__version__, pd.__version__, np.__version__)
```

```python
df = pd.read_csv('00-1-shipman-confirmed-victims-x.csv')
df
```

## Scatter Plot

```python
plt.scatter(df['fractionalDeathYear'], df['Age']);
```

```python
man_age = df['Age'][df['gender']==1]
man_year = df['fractionalDeathYear'][df['gender']==1]
plt.scatter(man_year, man_age);
```

```python
woman = df[df['gender']==0]
plt.scatter(woman['fractionalDeathYear'], woman['Age'], s=10);
```

```python
plt.scatter(man_year, man_age, s=10, c='blue', alpha=.5)
plt.scatter(woman['fractionalDeathYear'], woman['Age'], s=10, c='red', alpha=.5);
plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Figure 0.1. Age vs year of death of Shipman\'s victims');
```

## Plotting with selected columns

```python
data = df[['fractionalDeathYear', 'Age', 'gender2']]
data.head()
```

```python
sns.jointplot(data=data, x='fractionalDeathYear', y='Age', marginal_kws=dict(bins=25), joint_kws=dict(s=20));
```

```python
g = sns.jointplot(data=data, x='fractionalDeathYear', y='Age', hue='gender2',  joint_kws=dict(legend=False));
# g.legend(loc='center')
```

```python
sns.set() # various default settings by sns
g = sns.jointplot(data=data, x='fractionalDeathYear', y='Age', marginal_kws=dict(bins=25));
```

```python
g.savefig('Fig0.1.png')
```

```python
g = sns.pairplot(data, hue='gender2', height=3, aspect=1.8);
g.savefig('fig0.1.pairplot.png')
```
